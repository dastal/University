```python
from io import StringIO
import requests

def load_dataset(url):
    r = requests.get(url)
    data = r.content.decode('utf8')
    df = pd.read_csv(StringIO(data), sep='\t')
    df.columns = ['tweet', 'label']
    return df

print(df.head(100))
```

```python
df_train_dev = load_dataset(url_train_dev)
df_test = load_dataset(url_test)
```

```python
match=1
list_nc=[]
for test in df_test.label.unique():
  match=0
  for train in df_train_dev.label.unique():
    if train==test:
      match=1
  if match==0:
    print("Exists on test set, but not in data set----->", test)
    list_nc.append(test)
```

So we have labels here (mr,mn,eu,...) that we will not be able to predicte, because our data set is cannot have train on them

We split our data into inputs and outputs, it is important to notice that data should be shuffled first, so that we can have data of all labels


```python
train_lim=int(df_train_dev.shape[0]*0.9)
train_data=df_train_dev.sample(frac=1)

X_train = train_data.tweet
y_train = train_data.label

df_test_prev=df_test
for wrong_label in list_nc:
  df_test_prev = df_test_prev.drop(df_test_prev[df_test_prev.label == wrong_label].index)

X_test = df_test_prev.tweet
y_test = df_test_prev.label
```

As done in the example code, we contrast whether until now any data is missing

```python
match=1
for test in y_test.unique():
  match=0
  for train in df_train_dev.label.unique():
    if train==test:
      match=1
  if match==0:
    print("Exists on test set, but not in data set----->", test)
```

```python
print('Initial set shape: ', df_train_dev.shape)
print('Initial labels shape: ', df_train_dev.shape)
print('Training set shape: ', X_train.shape)
print('Training labels shape: ', y_train.shape)
print('Test set shape vs initial test set: ', X_test.shape, df_test.tweet.shape)
print('Test labels shape vs initial test set: ', y_test.shape,df_test.label.shape)
```

We see that in the filtering we did before, we lost nearly no data (only 14 values out of 13k)

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
y_train = label_encoder.fit_transform(y_train)
y_test = label_encoder.transform(y_test)
```

Let's now train the multinomial naive Bayes model using GridSearch cv

Since we had an array before with a counter of the number of tweets per country, we can use it as a prior for the model, we will compare this result with the obtained by default

```python
dist_tweets=tweets/sum(tweets)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import GridSearchCV

mnb_pipeline = Pipeline([
    ('vect', CountVectorizer()),
    ('tfidf', TfidfTransformer()),
    ('nb', MultinomialNB())
])

param_grid_mnb = {'vect__ngram_range': [(1, 1), (1, 2)],
             'nb__alpha': [0.5,1,1.5],
             'nb__class_prior': [None,dist_tweets]}

gs_mnb = GridSearchCV(mnb_pipeline, param_grid_mnb, cv=10, n_jobs=-1, verbose=3)
gs_mnb.fit(X_train, y_train)
```

```python
mnb_df = pd.DataFrame.from_dict(gs_mnb.cv_results_)
mnb_df.sort_values(by=["rank_test_score"])
```

Now let's apply the same to the SGD Classifier

```python
from sklearn.linear_model import SGDClassifier

sgd_pipeline = Pipeline([
    ('vect', CountVectorizer()),
    ('tfidf', TfidfTransformer()),
    ('sgd', SGDClassifier())
])

param_grid_sgd = {
    'vect__ngram_range': [(1, 1), (1, 2)],
    'sgd__loss' : ['hinge', 'log', 'squared_hinge'],  # loss function
    'sgd__alpha' : [0.0001, 0.001, 0.01, 0.1],        # regulatisation
    'sgd__early_stopping' : [True, False]             # early stopping
}

gs_sgd = GridSearchCV(sgd_pipeline, param_grid_sgd, cv=10, n_jobs=-1, verbose=3)
gs_sgd.fit(X_train, y_train)
```

```python
sgd_df = pd.DataFrame.from_dict(gs_sgd.cv_results_)
sgd_df.sort_values(by=["rank_test_score"])
```

Best result for the multinomial naive Bayes model:
parameters: alpha (for regularization): 0.5

```python
mnb_df.sort_values(by=["rank_test_score"]).head(1)
```

Best result for the SGD model:
Parameters: 
alpha (regularization): 0.0001, early stopping set to false and hinge loss function used  


```python
sgd_df.sort_values(by=["rank_test_score"]).head(1)
```

As we can see, the SGD classifier model performed siginficantly better with an 0.84 mean test score, against the moltinomial naive Bayes with an 0.67 mean test score.
GridSearchCV generates candidate values for the parameters from an grid of candidate parameters. It goes through all possible combinations from the given parameters and finds the best one.

Creating a confusion matrix to visualise which languages were likely to be confused with each other. The diagonal corresponds to correctly identified languages.

```python
from sklearn.metrics import confusion_matrix

y_pred = gs_sgd.best_estimator_.predict(X_test)

conf_matrix = confusion_matrix(y_test, y_pred)

print(conf_matrix)
```
